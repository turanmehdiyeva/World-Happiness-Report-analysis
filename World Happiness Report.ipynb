### Happiest country in the world

```python
result = pd.concat([data15.head(10),data16.head(10),data17.head(10),data18.head(10),data19.head(10)],axis=0)
result.reset_index(drop=True,inplace=True)
```

```python
res = result.groupby('Country').mean()
res[['Happiness Score']].sort_values(by='Happiness Score',ascending=False)
```

```python
countries = sorted(list(result['Country'].unique()))
countries
```

```python
plt.plot(countries,(res['Happiness Score']), color='brown')
plt.xticks(rotation = 'vertical')
plt.xlabel('Countries')
plt.ylabel('Average ranking')
plt.title('the Happiest Countries in the world'.upper())
plt.grid()
plt.show()
```

```python
import numpy as np
def absolute_value(val):
    a  = np.round(val/100.*res['Happiness Score'].sum(), 2)
    return a
plt.pie(res['Happiness Score'], labels=countries,
        autopct=absolute_value, shadow=True)

plt.axis('equal')
plt.show()
```

### Is overall happiness increasing or decreasing?

```python
data15['Year'] = 2015
data16['Year'] = 2016
data17['Year'] = 2017
data18['Year'] = 2018
data19['Year'] = 2019
```

```python
merged_data = pd.concat([data15,data16,data17,data18,data19],axis=0)
merged_data.reset_index(drop=True,inplace=True)
merged_data = merged_data[['Year','Happiness Score']]
```

```python
happiness = merged_data.groupby('Year').mean()
happiness
```

```python
years = sorted(list(merged_data['Year'].unique()))
years
```

```python
plt.plot(years,(happiness['Happiness Score']), color='brown')
plt.xlabel('Years')
plt.ylabel('Happiness')
plt.title('Changing happiness in the world through the years'.upper())
plt.grid()
plt.show()
```
